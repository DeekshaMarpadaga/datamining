```python
Student_df['Semester'].replace(to_replace=['F','S'],value=[1,2],inplace=True)
```

```python
Student_df['StageID'].replace(to_replace=['lowerlevel','MiddleSchool','HighSchool'],value=[1,2,3],inplace=True)
```

```python
Student_df['ParentAnsweringSurvey'].replace(to_replace=['Yes','No'],value=[True,False],inplace=True)
```

```python
Student_df['AbsencesOver7'].replace(to_replace=['Above-7','Under-7'],value=[True,False],inplace=True)
```

```python
Student_df['Final_Grades'].replace(to_replace=['L','M','H'],value=[0,1,2],inplace=True)
```

```python
X = Student_df.drop(['Final_Grades'], axis = 1)
y = Student_df['Final_Grades']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

```python
predicted
```

```python
model.score(X_test, y_test)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (20,10), dpi=300)
tree.plot_tree(model, max_depth = 20, feature_names = X.columns, class_names = {0: "Low", 1:"Mid",2:"High"}, filled=True)
plt.show()
```

```python
fi = model.feature_importances_ #feature importance array
fi = pd.Series(data = fi, index = X.columns) #convert to Pandas series for plotting
fi.sort_values(ascending=False, inplace=True) #sort descending
```

```python
fi
```

```python
plt.figure(figsize=(12, 20))
chart = sns.barplot(x=fi, y=fi.index, palette=sns.color_palette("BuGn_r", n_colors=len(fi)))
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm=confusion_matrix(y_test, predicted)
```

```python
matrix=sns.heatmap(cm,annot=True)
matrix.set_xlabel('Predicted Values')
matrix.set_ylabel('Actual Values ');
matrix.xaxis.set_ticklabels(['Low','Mid','High'])
matrix.yaxis.set_ticklabels(['Low','Mid','High'])
plt.show()
```
